```python
# 가족 병력과 개인 병력 간의 확률 계산 함수
def calculate_family_to_personal_risk(df_NaNx, family_col, personal_col):
    # 가족 병력 유형 확인
    family_conditions = df_NaNx[family_col].unique()
    
    # 결과를 저장할 딕셔너리 초기화
    risk_probabilities = {}
    
    # 각 가족 병력 조건에 대해 확률 계산
    for condition in family_conditions:
        if condition != "x":  # 가족 병력이 없는 경우는 제외
            # 가족 병력이 있는 경우 필터링
            family_condition_df_NaNx = df_NaNx[df_NaNx[family_col] == condition]
            
            # 개인 병력이 동일한 경우 비율 계산
            total_cases = len(family_condition_df_NaNx)
            matching_cases = len(family_condition_df_NaNx[family_condition_df_NaNx[personal_col] == condition])
            
            # 확률 계산
            risk_probabilities[condition] = matching_cases / total_cases if total_cases > 0 else 0
    
    return risk_probabilities

# 가족 병력이 개인 병력으로 이어지는 확률 계산
risk_probabilities = calculate_family_to_personal_risk(df_NaNx, "family_medical_history", "medical_history")

# 결과 확인
risk_probabilities
```

```python
# 흡연여부가 심장질환과 고혈압에 미치는 영향
def analyze_disease_by_smoking(df_NaNx, family_col, smoker_col, personal_col, exclude_conditions, target_conditions):
    # 가족 병력에서 특정 조건을 제외한 데이터 필터링
    filtered_df_NaNx = df_NaNx[~df_NaNx[family_col].isin(exclude_conditions)]
    
    # 흡연 여부 그룹화
    smoker_groups = filtered_df_NaNx.groupby(smoker_col)
    
    # 흡연 여부별 목표 병력 조건 비율 계산
    disease_stats = {}
    for group, group_df_NaNx in smoker_groups:
        group_stats = {}
        total = len(group_df_NaNx)
        
        for condition in target_conditions:
            condition_cases = len(group_df_NaNx[group_df_NaNx[personal_col] == condition])
            group_stats[condition] = condition_cases / total if total > 0 else 0
        
        disease_stats[group] = group_stats
    
    return disease_stats

# 분석 설정
exclude_conditions = ["Heart disease", "High blood pressure"]  # 가족 병력에서 제외 조건
target_conditions = ["Heart disease", "High blood pressure"]  # 분석할 개인 병력 조건

# 분석 실행
disease_by_smoking = analyze_disease_by_smoking(
    df_NaNx, "family_medical_history", "smoker", "medical_history", exclude_conditions, target_conditions
)

# 결과 확인
disease_by_smoking
```

```python
# bmi지수 별 당뇨여부
import numpy as np

# BMI와 당뇨병 간의 관계 분석 함수
def analyze_bmi_and_diabetes(df_NaNx, bmi_col, medical_col, target_condition="Diabetes", bins=None):
    # 당뇨병 여부 필터링
    df_NaNx["diabetes"] = df_NaNx[medical_col].apply(lambda x: 1 if x == target_condition else 0)
    
    # BMI 구간 생성 (선택적)
    if bins:
        df_NaNx["bmi_category"] = pd.cut(df_NaNx[bmi_col], bins=bins, labels=False, include_lowest=True)
        grouped = df_NaNx.groupby("bmi_category")
        
        # 각 BMI 그룹에서 당뇨병 비율 계산
        diabetes_by_bmi = grouped["diabetes"].mean()
        return diabetes_by_bmi
    else:
        # 상관계수 계산
        correlation = np.corrcoef(df_NaNx[bmi_col], df_NaNx["diabetes"])[0, 1]
        return correlation

# BMI와 당뇨병 여부 관계 분석 (BMI를 연속 변수로 분석)
bmi_diabetes_correlation = analyze_bmi_and_diabetes(df_NaNx, "bmi", "medical_history")

# BMI를 구간화하여 분석
bmi_bins = [0, 18.5, 25, 30, 35, 40, np.inf]  # BMI 구간 정의
bmi_diabetes_by_category = analyze_bmi_and_diabetes(df_NaNx, "bmi", "medical_history", bins=bmi_bins)

# 결과 확인
bmi_diabetes_correlation, bmi_diabetes_by_category
```

```python
# 운동 빈도와 질병 발생률 간의 관계 분석
def analyze_exercise_and_disease(df_NaNx, exercise_col, medical_col):
    # 운동 빈도별 그룹화
    exercise_groups = df_NaNx.groupby(exercise_col)
    
    # 각 그룹의 질병 발생률 계산
    disease_stats = {}
    for group, group_df_NaNx in exercise_groups:
        total = len(group_df_NaNx)
        has_disease = len(group_df_NaNx[group_df_NaNx[medical_col] != "None"])  # 병력이 있는 경우 필터링
        disease_stats[group] = has_disease / total if total > 0 else 0
    
    return disease_stats

# 운동 빈도와 질병 발생률 간의 관계 분석
exercise_disease_relationship = analyze_exercise_and_disease(df_NaNx, "exercise_frequency", "medical_history")

# 결과 확인
exercise_disease_relationship
```
